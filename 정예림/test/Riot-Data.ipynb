```python
from riotwatcher import LolWatcher, ApiError
from pprint import pprint
from datetime import datetime, timedelta
import os
```

```python
f = open("api.txt", "r")
api = f.readline()
```

```python
lol_watcher = LolWatcher(api)

my_region = 'kr'
```

```python
me = lol_watcher.summoner.by_name(my_region, 'hide on bush')

print(me)
```

```python
my_ranked_stats = lol_watcher.league.by_summoner(my_region, me['id'])
pprint(my_ranked_stats)
```

```python
my_matches = lol_watcher.match.matchlist_by_account(my_region, me['accountId'])

last_match = my_matches['matches'][0]
match_detail = lol_watcher.match.by_id(my_region, last_match['gameId'])

print('KDA\n')
assists = match_detail['participants'][0]['stats']['assists']
kills = match_detail['participants'][0]['stats']['kills']
deaths = match_detail['participants'][0]['stats']['deaths']
print( (assists + kills) / deaths )
print('\n');

print('0-10분 데미지\n')
print(match_detail['participants'][0]['timeline']['damageTakenPerMinDeltas']['0-10'])
print('\n');

print('0-10분 획득 골드\n')
print(match_detail['participants'][0]['timeline']['goldPerMinDeltas']['0-10'])
print('\n');

print('0-10분 획득 경험치\n')
print(match_detail['participants'][0]['timeline']['xpPerMinDeltas']['0-10'])
print('\n');

#pprint(match_detail)
```
